```python
import os
import numpy as np
import cv2
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Load data function
def load_data(directory):
    images = []
    labels = []
    for filename in os.listdir(directory):
        if filename.endswith(".jpg"):
            img = cv2.imread(os.path.join(directory, filename))
            img = cv2.resize(img, (150, 150))  
            images.append(img)
            if "metal" in filename:
                labels.append(1)  
            elif "glass" in filename:
                labels.append(2)
            elif "paper" in filename:
                labels.append(3)
            elif "plastic" in filename:  # Handle plastic images
                labels.append(4)
            else:
                labels.append(0)  
    return np.array(images), np.array(labels)

# Load data from "metal" directory
metal_images, metal_labels = load_data("metal")

# Load data from "glass" directory
glass_images, glass_labels = load_data("glass")

# Load data from "paper" directory
paper_images, paper_labels = load_data("paper")

# Load data from "plastic" directory
plastic_images, plastic_labels = load_data("plastic")

# Combine datasets
all_images = np.concatenate((metal_images, glass_images, paper_images, plastic_images), axis=0)
all_labels = np.concatenate((metal_labels, glass_labels, paper_labels, plastic_labels), axis=0)

# Update labels for trash images
all_labels[np.where(all_labels == 0)] = 5  # Update labels for trash to a different value, e.g., 5

# Split data into training and testing sets
train_images, test_images, train_labels, test_labels = train_test_split(all_images, all_labels, test_size=0.2, random_state=42)

# Data Augmentation
datagen = ImageDataGenerator(
    rotation_range=20,  # randomly rotate images in the range (degrees, 0 to 180)
    width_shift_range=0.2,  # randomly shift images horizontally (fraction of total width)
    height_shift_range=0.2,  # randomly shift images vertically (fraction of total height)
    shear_range=0.2,  # randomly apply shearing transformations
    zoom_range=0.2,  # randomly zoom in/out
    horizontal_flip=True,  # randomly flip images horizontally
    vertical_flip=True  # randomly flip images vertically
)

datagen.fit(train_images)

# Design the model
model = Sequential([
    Conv2D(filters=32, kernel_size=3, padding='same', activation='relu', input_shape=(150, 150, 3)),
    MaxPooling2D(pool_size=2),
    Conv2D(filters=64, kernel_size=3, padding='same', activation='relu'),
    MaxPooling2D(pool_size=2),
    Conv2D(filters=128, kernel_size=3, padding='same', activation='relu'),
    MaxPooling2D(pool_size=2),
    Conv2D(filters=128, kernel_size=3, padding='same', activation='relu'),
    MaxPooling2D(pool_size=2),
    Flatten(),
    Dense(6, activation='softmax')  # Changed output to 6 classes to include plastic
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model with augmented data
model.fit(datagen.flow(train_images, train_labels, batch_size=32), epochs=60, validation_data=(test_images, test_labels))

# Evaluate the model
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f"Model accuracy on test data: {test_acc}")

# Save the model
model.save('metal_glass_paper_plastic_model_with_augmentation.h5')
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

# Load the trained model
model = load_model('metal_glass_paper_plastic_model_with_augmentation.h5')

# Function to predict the type of the object in the photo
def predict_object(image_path):
    # Load and preprocess the image
    img = cv2.imread(image_path)
    img = cv2.resize(img, (150, 150))
    img = np.expand_dims(img, axis=0)  # Add batch dimension

    # Make prediction
    predictions = model.predict(img)

    # Get the predicted class and confidence
    class_id = np.argmax(predictions)
    confidence = predictions[0][class_id] * 100
    
    # Map class ID to class name
    class_names = {0: 'Unknown', 1: 'Metal', 2: 'Glass', 3: 'Paper', 4: 'Plastic', 5: 'Trash'}
    predicted_class = class_names[class_id]

    return predicted_class, confidence

# Function to capture photo from webcam
def capture_photo():
    # Initialize the webcam
    cap = cv2.VideoCapture(0)
    
    # Check if the webcam is opened correctly
    if not cap.isOpened():
        print("Error: Unable to open webcam")
        return

    # Capture a frame
    ret, frame = cap.read()
    
    # Save the captured frame as temp.jpg
    cv2.imwrite('temp.jpg', frame)

    # Release the webcam
    cap.release()

# Capture a photo
capture_photo()

# Predict the type of object in the captured photo
image_path = 'temp.jpg'
predicted_class, confidence = predict_object(image_path)

# Display the prediction
print(f"Predicted class: {predicted_class}")
print(f"Confidence: {confidence:.2f}%")
```
